```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

users = pd.read_csv("data/users.csv")
orders = pd.read_csv("data/orders.csv")
clicks = pd.read_csv("data/clicks.csv")
ads = pd.read_csv("data/ads.csv")
```

```python
users['gender'].value_counts().plot(kind='bar')
plt.title("ç”¨æˆ·æ€§åˆ«åˆ†å¸ƒ")
plt.show()
```

## 3. è®¢å•è¶‹åŠ¿åˆ†æž

```python
orders['order_time'] = pd.to_datetime(orders['order_time'])
orders['order_date'] = orders['order_time'].dt.date
orders.groupby('order_date')['price'].sum().plot()
plt.title("æ¯æ—¥é”€å”®é¢")
plt.show()
```
